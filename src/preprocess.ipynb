```python

import os
import numpy as np
import pandas as pd
import gzip


def read_DNA(dirname):
    count = 0
    nDNA=0
    DNA_data=[]
    #read DNA data
    for idname in os.listdir(dirname):
        # list all the ids 
        if(count > 1000):
            break
        if idname.find("-") != -1:
            idpath = dirname +"/" + idname

            # all the files in each id directory
            for filename in os.listdir(idpath):
                # check the gene file
                if filename.find("-") != -1:

                    filepath = idpath + "/" + filename

                    with gzip.open(filepath, 'rt') as f:
                        df = pd.read_csv(f,sep="\t", header=None)
                    # columns = ["DNA_ID", "read_count"]
                    #if count ==0:
                        # get the DNA_IDs 
                        
                    if nDNA != 0 and nDNA != len(df.index) :
                        continue
                        print (filename + ", "+ idname +" mismatch")
                        print (nDNA, len(df.index))
                        continue
                    nDNA = len(df.index)
                    
                    df = df.iloc[0:1000,:]
                    
                    if count != 0:
                        DNA_IDs = df.iloc[:,0].values.tolist()
                    #    with conn.cursor() as cursor:
                    #        sql = "alter table `Gene` add column `"+idname+"` double"
                    #        print (sql)
                    #        cursor.execute(sql)

                    #df.columns = ['DNA_ID',idname]

                    
                    id_DNA_read_counts = [idname] + df.iloc[:,1].values.tolist()
                    DNA_data.append(id_DNA_read_counts)
                    #print (count)

                    count +=1

    columns = ["file_id"] + DNA_IDs
    df = pd.DataFrame(DNA_data, columns=columns)
    return df

dirname='../data/gene'
df=read_DNA(dirname)
df.to_csv('../data/GeneExpression.csv')
df.head(10)
```
